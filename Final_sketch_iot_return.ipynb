```python
import requests
import time
```

```python
def read_values_from_thingspeak(api_key, channel_id, results=2):
    try:
        url = f"https://api.thingspeak.com/channels/{channel_id}/feeds.json?api_key={api_key}&results={results}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            for entry in data['feeds']:
                print(f"Entry ID: {entry['entry_id']}")
                print(f"Field 1: {entry['field1']}")
                value=entry['field1']
                return value
        else:
            print(f"Error: Unable to fetch content. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
def grayToBinary():
    api_key = "I86SI9SND03A0SZK"
    channel_id = 2379087
    gray_code=read_values_from_thingspeak(api_key, channel_id, results=2)
    # storing length of gray in n
    n = len(gray_code)
    
    # starting with an empty string, that will be our answer.
    binary = ""
    
    # adding MSB of gray to answer as it is.
    binary += gray_code[0]
    
    # iterating till the length of gray.
    for i in range(1, n):
    
        # If the digit is 0, copy the previous digit to answer,
        if gray_code[i] == '0':
            binary += binary[i - 1]
        else:
            if binary[i - 1] == '0':
                binary += '1'
            else:
                binary += '0'
    return binary
```
